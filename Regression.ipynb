```python
y_pred=reg.predict(x_test)
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

from sklearn.metrics import mean_absolute_error, mean_squared_error
Here you‚Äôre importing two heavy hitters from scikit-learn‚Äôs toolbox to judge how good (or bad) your model‚Äôs predictions are:

Mean Absolute Error (MAE): The average absolute difference between what your model predicted and the real values. This one is straightforward ‚Äî how far off, on average, is your prediction?

Mean Squared Error (MSE): Similar to MAE but punishes bigger errors more heavily by squaring the differences. It‚Äôs like saying ‚ÄúBig mistakes? Not on my watch.‚Äù

```python
mse = mean_squared_error(y_test,y_pred)
mae= mean_absolute_error(y_test,y_pred)
rmse = np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
```

mse = mean_squared_error(y_test, y_pred)
Calculates the average of squared differences between actual values (y_test) and your predictions (y_pred). Squaring means big mistakes hurt way more, so it‚Äôs sensitive to outliers.
Lower is better. MSE = 0 means perfect predictions ‚Äî but come on, that‚Äôs rare.

mae = mean_absolute_error(y_test, y_pred)
Averages the absolute difference between actuals and predictions ‚Äî treats all errors equally. Gives you a more ‚Äúhuman-friendly‚Äù sense of average error. Also, lower = better.

rmse = np.sqrt(mse)
Root Mean Squared Error is just the square root of MSE ‚Äî this brings the error back to the original units (instead of squared units). It‚Äôs easier to interpret because it‚Äôs on the same scale as your target variable.

```python
residuals=y_test-y_pred
print(residuals)
```

residuals = y_test - y_pred
You calculated the difference between the actual values and what your model predicted. Basically, residuals are your model‚Äôs oops moments ‚Äî how far off it was on each prediction.

print(residuals)
You're now looking at those errors laid bare. Each value shows whether your prediction was too high (negative residual) or too low (positive residual).

Why residuals matter:
If residuals look like random noise (no pattern), your model is probably doing its job right.

But if residuals show some pattern (like a curve or trend), your model might be missing some hidden magic in the data ‚Äî time to rethink your features or model.




```python
sns.displot(residuals,kind='kde')
```

What you‚Äôre doing here:
You‚Äôre plotting a Kernel Density Estimate (KDE) of your residuals ‚Äî basically a smooth curve showing how residuals spread out.

Ideally, residuals should look like a normal (bell-shaped) curve centered at zero. That means errors are mostly small and symmetrically distributed around zero ‚Äî model behaving like a well-trained.
What to watch for:
Symmetry & Centering: If the curve peaks at zero and is symmetric, nice job!

Skewed or off-centered: Means bias ‚Äî model consistently over or underpredicting.

Multiple peaks or weird shapes: Uh-oh, your model might be missing key patterns or suffering from weird noise.

```python
plt.scatter(y_pred,residuals)
```

What to watch out for:
No pattern, random cloud: Your model‚Äôs errors are likely independent of predicted values ‚Äî good sign!

Clear patterns (like a funnel or curve): Uh-oh, that means heteroscedasticity (errors‚Äô variance changes with prediction) or model missing something ‚Äî not ideal!

Clusters or trends: Could indicate model bias, bad fit, or non-linearity. Time to level up the model.



```python
import statsmodels.api as sm
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()


```

import statsmodels.api as sm
Pulling in the big gun for statistical modeling. Statsmodels is the OG library if you want full control and detailed regression results ‚Äî no sugarcoating, just cold hard stats.

x = sm.add_constant(x)
This adds the intercept term (the constant) to your features matrix. Unlike scikit-learn, statsmodels won‚Äôt automatically include it, so you gotta explicitly add this column of 1s. Otherwise, your regression line will be forced through the origin ‚Äî yikes.

model = sm.OLS(y, x).fit()
You‚Äôre telling statsmodels:
‚ÄúHere‚Äôs my dependent variable y and independent variables x. Fit me an Ordinary Least Squares (OLS) linear regression model.‚Äù

```python
model.summary()
```

model.summary()! This command spits out the full regression report like a boss, giving you the inside scoop on how your model is flexing.

What to expect in the summary:
R-squared & Adj. R-squared:
How much of the variation in your dependent variable your model explains. Closer to 1 = better fit. Adjusted version punishes you for too many useless features.

Coefficients:
The effect size of each feature on the target variable. Positive means it pushes your target up; negative means it drags it down.

Standard Errors:
How ‚Äúsure‚Äù the model is about each coefficient. Smaller = more confidence.

t-statistics & p-values:
Hypothesis test results. p-values < 0.05 usually mean that feature actually matters ‚Äî no BS.

F-statistic:
Overall significance of the model ‚Äî are your predictors working together or just noise?

Diagnostic stuff:
Durbin-Watson for autocorrelation, Omnibus test for normality, Jarque-Bera test ‚Äî basically, how clean and trustworthy your model assumptions are.
